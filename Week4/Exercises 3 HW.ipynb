```python
def finiteDifference(x, y):
    dy = diff(y)
    dx = diff(x)
    dydx1 = zeros(y.shape, float) 
    dydx1[:-1] = dy/dx
    dydx1[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])
    return dydx1
```

b)
-

```python
x = arange(0,10.1,.001)
y = x**5
s = finiteDifference(x, y)
print s
real = 5*x**4
plot(x,real, linewidth = 3)
plot(x,s, linestyle = "--", linewidth = 5, color = 'r')
show()
#plots show that they are the same
```

c)
-

```python
percentError = 100*abs(s[1:] - real[1:])/real[1:]
print percentError
semilogy(x[1:], percentError, label = "percent Error")
legend(loc = "upper right")
ylabel("percent error")
show()
```

```python
def fourPtFiniteDiff(x,y):
    h = .1#space of x values on x axis
    dydx = zeros(y.shape, float)
    for i in range(2, len(y)-2):
        dydx[i] = (y[i - 2] - 8*y[i - 1] + 8*y[i + 1] - y[i + 2])/(12*h)#formula from assignment
    return dydx
    
```

```python
a = arange(0, 10.1, .001)
b = x**5
d = fourPtFiniteDiff(a,b)
print d
plot(x,d)
```

```python
percentError = 100*abs(d[2:] - real[1:])/real[1:]
print percentError
semilogy(x[1:], percentError, label = "percent Error")
legend(loc = "upper right")
ylabel("percent error")
show()
#this only works when h from the function defined above as fourPtFiniteDiff(x,y) is .001
```

```python
x = linspace(0,2*pi, 100)
y = sin(x) + 0.1*random.random(size=x.shape)
plot(x,y, label = "original data")
finite = finiteDifference(x,y)
plot(x,finite, label = "2pt-forward diff")
four = fourPtFiniteDiff(x,y)
plot(x,four, label = "4pt-centerent diff")
plot(x, cos(x), label = "analytical derivative")
legend(loc = "lower left")
ylabel("Noisy data and derivatives")
xlabel("x")
show()
```
