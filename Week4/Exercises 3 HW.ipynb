```python
def Temp(Resistance):
    """this function takes any value of resistance and returns the interpolated temp"""
    A = 1/10 - (B + math.log(3.239e6)**2 * C)*math.log(3.239e6)
    B = (1/11 - 1/10)/(math.log(3.118e6) - math.log(3.239e6)) - C*(math.log(3.118e6)**2 + math.log(3.118e6)*math.log(3.239e6) + math.log(3.239e6)**2)
    C = (((1/12 - 1/10)/(math.log(3.004e6) - math.log(3.239e6)) - (1/11 - 1/10)/(math.log(3.118e6) - math.log(3.239e6)))/(math.log(3.004e6) - math.log(3.118e6)) * 1 / (math.log(3.239e6) + math.log(3.118e6) + math.log(3.004e6))
    temp = 1 /(A + B*math.log(Resistance) + C*math.log(Resistance)**3)
    T = temp - 272.15
    return T
```

```python
def finiteDifference(x, y):
    dy = diff(y)
    dx = diff(x)
    dydx1 = zeros(y.shape, float) 
    dydx1[:-1] = dy/dx
    dydx1[-1] = (y[-1] - y[-2])/(x[-1] - x[-2])
    return dydx1
```

```python
x = arange(0,10.1,.001)
y = x**5
s = finiteDifference(x, y)
print s
real = 5*x**4
plot(x,real, linewidth = 1)
plot(x,s, linestyle = "--")
show()
#plots show that they are the same
```

```python
x = linspace(0,2*pi, 100)
y = sin(x) + 0.1*random.random(size=x.shape)
plot(x,y)
title("noisy sinusoidal")
show()
```

```python
def fourPtFiniteDiff(x,y):
    h = .1
    dydx = zeros(y.shape, float)
    for i in range(2, len(y)-2):
        dydx[i] = (y[i - 2] - 8*y[i - 1] + 8*y[i + 1] - y[i + 2])/(12*h)
    return dydx
    
```

```python
a = arange(0, 10.1, .001)
b = x**5
d = fourPtFiniteDiff(a,b)
print d
plot(x,d)
```

```python
percentError = 100*abs(d[2:] - real[2:])/real[2:]
print percentError
semilogy(x[2:], percentError, label = "percent Error")
legend(loc = "upper right")
ylabel("percent error")
show()
```

```python
x = linspace(0,2*pi, 100)
y = sin(x) + 0.1*random.random(size=x.shape)
plot(x,y, label = "original data")
finite = finiteDifference(x,y)
plot(x,finite, label = "2pt-forward diff")
four = fourPtFiniteDiff(x,y)
plot(x,four, label = "4pt-centerent diff")
plot(x, cos(x), label = "analytical derivative")
legend(loc = "lower left")
ylabel("Noisy data and derivatives")
xlabel("x")
show()
```
