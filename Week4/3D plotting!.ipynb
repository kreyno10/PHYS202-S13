Electric Potential 
--

$V(x,y) = \frac{kq}{\sqrt{x^2 + (y - d)^2}} - \frac{kq}{\sqrt{x^2+ (y + d)^2}}$

```python
#make xy grid 
dx = 0.06
dy = 0.06
x = arange(-2,2,dx)
y = arange(-2,2,dy)
X,Y = meshgrid(x,y)
#make the function V(x,y)
k = 1; q = 1; d = 1
V = (k*q/ (X**2 + (Y - d)**2)**(1/2.)) - (k*q/ (X**2 + (Y + d)**2)**(1/2.))
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, linewidth=0, cmap=cm.rainbow)
cb = fig.colorbar(p, shrink=0.5)
```

```python
levels = array([8, 4, 2, 1, 0.5, 0.25, 0, -.25, -.5, -1, -2, -4, -8])
contour(X,Y,V, levels, linewidths = 10, cmap=cm.rainbow)
show()
```

```python
#try it on your own!
dx = .051
dy = .051
x = arange(-6,6,dx)
y = arange(-6,6,dy)
X,Y = meshgrid(x,y)
G = e**(-X**2 - Y**2)

from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X, Y, G, rstride=1, cstride=1, linewidth=0, cmap=cm.rainbow)
cb = fig.colorbar(p, shrink=0.5)
cse = ax.contour(X,Y,G, array([0.,.2,.4,.6,.8]), zdir = 'x', offset=-3)
```

```python
levels = array([8, 4, 2, 1, 0.5, 0.25, 0, -.25, -.5, -1, -2, -4, -8])
contour(X,Y,G, levels, linewidths = 1, cmap=cm.rainbow)
show()
```
