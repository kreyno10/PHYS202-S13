Electric Potential 
--

$V(x,y) = \frac{kq}{\sqrt{x^2 + (y - d)^2}} - \frac{kq}{\sqrt{x^2+ (y + d)^2}}$

```python
#make xy grid 
dx = 0.06
dy = 0.06
x = arange(-2,2,dx)
y = arange(-2,2,dy)
X,Y = meshgrid(x,y)
#make the function V(x,y)
k = 1; q = 1; d = 1
V = (k*q/ (X**2 + (Y - d)**2)**(1/2.)) - (k*q/ (X**2 + (Y + d)**2)**(1/2.))
print V
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection = '3d')
p = ax.plot_surface(X, Y, V, rstride=1, cstride=1, linewidth=0, cmap=cm.rainbow)
cb = fig.colorbar(p, shrink=0.5)
```

```python
levels = array([8, 4, 2, 1, 0.5, 0.25, 0, -.25, -.5, -1, -2, -4, -8])
contour(X,Y,V, levels, linewidths = 10, cmap=cm.rainbow)
show()
```

```python
def pointPotential(x,y,q,posx,posy):
    #this returns the potential of this point
    k = 1
    V = (k*q/(x**2 + (y - (posx**2 + posy**2)**(1/2.))**2))
    return V
```

```python
x = 1.; y = 1
V11 = pointPotential(x,y,1e-9,0.,0.) 
print V11

```

```python
x,y = meshgrid(arange(-2.,2.,0.06),arange(-2.,2.,0.06)) 
Vxy = pointPotential(x,y,1e-9,0.,0.) 
print Vxy
```

```python
def dipolePotential(x,y,q,d):
    #this returns the potntial due to two points
    V = (k*q/ (x**2 + (y - d)**2)**(1/2.)) - (k*q/ (x**2 + (y + d)**2)**(1/2.))
    return V
  
```

```python
x = 1.; y = 1
Vdp = dipolePotential(x,y,1e-9,2e-9) 
print Vdp
```

```python
x,y = meshgrid(arange(-2.,2.,0.06),arange(-2.,2.,0.06)) 
V22 = dipolePotential(x,y,1,1) 
print V22
from mpl_toolkits.mplot3d import Axes3D
fig = figure()
ax = fig.gca(projection = '3d')
p1 = ax.plot_surface(x, y, V22, rstride=1, cstride=1, linewidth=0, cmap=cm.rainbow)
cb = fig.colorbar(p1, shrink=0.5)
```
