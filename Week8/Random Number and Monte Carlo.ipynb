CDF Example
--

```python
x = arange(0.,pi+0.001,0.01)
f = lambda x: sin(x)/2
F = lambda x: (cos(0)-cos(x))/2

plot(x,f(x),label="True Distribution",lw=3)
plot(x, F(x),label="Cumulative function",lw=3)

xlim(0,3.5)
ylim(0,1.1)

legend(loc=0)
show()
        
```

```python
#number of samples
N = 10000
#N random numbers between 0 and 1
z = numpy.random.random_sample(N);
#inverse CDF
Finv = lambda z: arccos(cos(0) - 2*z)
#sampled x values
xsamples = Finv(z);
#how do they compare to original?
hist(xsamples,50,normed=True,label="Sampled distribution")
plot(x,f(x),label="True distribution",lw=3,color="red")
xlim(0,pi)
ylim(0,0.7)
legend(loc=0)
show()
```

```python
N = 10000
def distribute1D(x, prob, N):
    sum_dist = np.cumsum(prob)
    normal = sum_dist/sum(prob)
    z = zeros(N, float)
    for i in range(N):
        z[i] = random.uniform(0,1)
    index = zeros(N, int)
    for i in range(N):
        index[i] = np.searchsorted(normal, z[i])
    value = zeros(N, float)
    for i in range(N):
        value[i] = x[index[i]]
    return value
    
```

```python
x = arange(0.,1000+0.001,0.01)
l = 1
f = lambda x: l*exp(-l*x)
distribution = []
for i in x:
    distribution.append(f(i))
test = distribute1D(x, distribution, 10)
print test
```
