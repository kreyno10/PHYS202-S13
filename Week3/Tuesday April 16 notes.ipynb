Using  Matplotlib to make pretty graphs

```python
#simple line plot
t = arange(0.0,1.01, .01)
c = cos(4*pi*t)
s = sin(4*pi*t)
```

```python
plot(t,s,color="blue", linewidth=2.5, label="$sin(4{\pi}t)$")
plot(t,c, color="red", linewidth=2.5, label="$cos(4{\pi}t)$")
xlabel("time in (s)", fontsize = 12)
ylabel("voltage in (mV)", fontsize = 12)
title("my cool plot", fontsize =20)
legend()
grid(True)
savefig("simple_plot.pdf")
show()
```
