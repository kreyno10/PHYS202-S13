```python
def pointField(x,y,q,Xq,Yq):
    k = 8.99 *10**9
    Ex = k*q*(x - Xq) / sqrt((x-Xq)**2 + (y - Yq)**2)
    Ey = k*q*(y-Yq) / sqrt((x-Xq)**2 + (y - Yq)**2)
    E = [Ex, Ey]
    return E
```
